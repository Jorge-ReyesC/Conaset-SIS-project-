# *Grafico De Tortas*

```python
import seaborn as sns
import matplotlib.pyplot as plt
desfase=[0.1,0]
plt.pie(df_genero_accidentes['Fallecidos'],
        labels=df_genero_accidentes['Genero'],
        colors=['w','y'],
        shadow=False,
        explode=desfase,
        autopct='%.1f%%')
plt.axis("equal")
plt.show()
```

# Interpretacion Del Grafico:

### Podemos observar en un grafico de tortas que el 79.4% De fallecidos por siniestros de transito en el año 2020 fueron hombres y le 20.6% restante corresponde a mujeres.

# *Grafico De Barras*

```python
df_accidentes.groupby('Tipo__CONA').count()["Fallecidos"].plot(kind='barh')
plt.title('Causa y Fallecidos por siniestros de transito',fontsize=20)
plt.xlabel('Fallecidos',fontsize=15)
plt.ylabel('Causas De Accidentes',fontsize=15)
```

# Interpretacion:

***Se Puede observar las causas las cuales mas se producen accidentes que en este caso son por choque y colision donde se ven un numero disparado de fallecidos en comparacion por atropello o caida***
